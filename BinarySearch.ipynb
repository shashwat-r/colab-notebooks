```python
# Create Binary Search Variants

binary_search_variants = []
LOOP_LIMIT = 100 # adjusted based on linear search iteration limit
def LOOP_LIMIT_RANGE(): return range(1,LOOP_LIMIT+1)
def HANDLE_LOOP_LIMIT_REACHED(loop_number):
    if loop_number >= LOOP_LIMIT: raise Exception('Max iterations reached')

description = """
Dummy Implementation - Baseline
Returns that the element isn't present
Will be true more than 50% times given the test design

Fails whenever an element is present
"""
def binary_search(arr, element):
    idx = -1
    return idx

binary_search_variants.append((binary_search, description))


description = """
First Implementation
Left and Right Indices are inclusive
Midpoint is matched or skipped over

Fails everytime a non existing element is requested - goes into infinite loop
"""
def binary_search(arr, element):
    idx = -1
    left_idx = 0
    right_idx = len(arr) - 1

    for loop_number in LOOP_LIMIT_RANGE():

        mid_idx = (left_idx + right_idx) // 2
        if arr[mid_idx] == element:
            idx = mid_idx
            break
        elif arr[mid_idx] < element:
            left_idx = mid_idx + 1
        else:
            right_idx = mid_idx - 1

        HANDLE_LOOP_LIMIT_REACHED(loop_number)

    return idx

binary_search_variants.append((binary_search, description))


description = """
Added breaking condition to handle missing elements

Fails only when list is empty
"""
def binary_search(arr, element):
    idx = -1
    left_idx = 0
    right_idx = len(arr) - 1

    for loop_number in LOOP_LIMIT_RANGE():

        mid_idx = (left_idx + right_idx) // 2
        if arr[mid_idx] == element:
            idx = mid_idx
            break
        elif arr[mid_idx] < element:
            left_idx = mid_idx + 1
        else:
            right_idx = mid_idx - 1
        if left_idx > right_idx:
            break

        HANDLE_LOOP_LIMIT_REACHED(loop_number)

    return idx

binary_search_variants.append((binary_search, description))


description = """
Moved the missing element breaking condition to top of the loop.

Passes all tests.
"""
def binary_search(arr, element):
    idx = -1
    left_idx = 0
    right_idx = len(arr) - 1

    for loop_number in LOOP_LIMIT_RANGE():

        if left_idx > right_idx:
            break
        mid_idx = (left_idx + right_idx) // 2
        if arr[mid_idx] == element:
            idx = mid_idx
            break
        elif arr[mid_idx] < element:
            left_idx = mid_idx + 1
        else:
            right_idx = mid_idx - 1

        HANDLE_LOOP_LIMIT_REACHED(loop_number)

    return idx

binary_search_variants.append((binary_search, description))


"""
Updated the breaking condition to break when indices are equal as well.

Fails when left and right indices get values of the correct element.
Passes when left and right indices are different and lead to the midpoint being the correct element.
"""
def binary_search(arr, element):
    idx = -1
    left_idx = 0
    right_idx = len(arr) - 1

    for loop_number in LOOP_LIMIT_RANGE():

        if left_idx >= right_idx:
            break
        mid_idx = (left_idx + right_idx) // 2
        if arr[mid_idx] == element:
            idx = mid_idx
            break
        elif arr[mid_idx] < element:
            left_idx = mid_idx + 1
        else:
            right_idx = mid_idx - 1

        HANDLE_LOOP_LIMIT_REACHED(loop_number)

    return idx

binary_search_variants.append((binary_search, description))


description = """
gets into list out of range when element greater than max element is requested
otherwise gets into infinite loop if element is not in array
"""
def binary_search(arr, element):
    idx = -1
    left_idx = 0
    right_idx = len(arr)

    for loop_number in LOOP_LIMIT_RANGE():

        mid_idx = (left_idx + right_idx) // 2
        if arr[mid_idx] == element:
            idx = mid_idx
            break
        elif arr[mid_idx] < element:
            left_idx = mid_idx + 1
        else:
            right_idx = mid_idx - 1

        HANDLE_LOOP_LIMIT_REACHED(loop_number)

    return idx

binary_search_variants.append((binary_search, description))


description = """
breaks when left and right indices both converge at midpoint, and not otherwise
"""
def binary_search(arr, element):
    idx = -1
    left_idx = 0
    right_idx = len(arr)

    for loop_number in LOOP_LIMIT_RANGE():
        if left_idx >= right_idx:
            break
        mid_idx = (left_idx + right_idx) // 2
        if arr[mid_idx] == element:
            idx = mid_idx
            break
        elif arr[mid_idx] < element:
            left_idx = mid_idx + 1
        else:
            right_idx = mid_idx - 1

        HANDLE_LOOP_LIMIT_REACHED(loop_number)

    return idx

binary_search_variants.append((binary_search, description))


description = """
gets into list out of range when element greater than max element is requested
"""
def binary_search(arr, element):
    idx = -1
    left_idx = 0
    right_idx = len(arr)

    for loop_number in LOOP_LIMIT_RANGE():
        if left_idx > right_idx:
            break
        mid_idx = (left_idx + right_idx) // 2
        if arr[mid_idx] == element:
            idx = mid_idx
            break
        elif arr[mid_idx] < element:
            left_idx = mid_idx + 1
        else:
            right_idx = mid_idx - 1

        HANDLE_LOOP_LIMIT_REACHED(loop_number)

    return idx

binary_search_variants.append((binary_search, description))


description = """
gets into list out of range when element greater than max element is requested
"""
def binary_search(arr, element):
    idx = -1
    left_idx = 0
    right_idx = len(arr)

    for loop_number in LOOP_LIMIT_RANGE():
        mid_idx = (left_idx + right_idx) // 2
        if arr[mid_idx] == element:
            idx = mid_idx
            break
        elif left_idx >= right_idx:
            break
        elif arr[mid_idx] < element:
            left_idx = mid_idx + 1
        else:
            right_idx = mid_idx - 1

        HANDLE_LOOP_LIMIT_REACHED(loop_number)

    return idx

binary_search_variants.append((binary_search, description))


idx = 1
func, desc = binary_search_variants[idx]
print('Index:', idx)
print(desc)
```

```python
import pandas as pd

inclusive_range = lambda l,r: range(l,r+1)
bool_to_num = lambda b: 1 if b else 0

def get_binary_search_inputs(array_count):
    binary_search_inputs = []

    if array_count < 1:
        binary_search_inputs.append({
            'input_array': [],
            'elements_info': {1: {'lower_bound': -1, 'upper_bound': -1}}
        })
        return binary_search_inputs

    new_element_count = array_count - 1
    for num in range(1 << new_element_count):

        # Create Input Array
        current_element = 2
        input_array = [current_element]
        for i in inclusive_range(1, new_element_count):
            mask = (1 << (new_element_count)) >> i
            to_increase = 1 if num & mask else 0
            current_element = current_element + 2 * to_increase
            input_array.append(current_element)

        # Decide Min and Max Elements to Search in Input Array
        min_element, max_element = 1, current_element+1 # less than the min(array) and more than max(array)

        # Create Upper and Lower Bound Dicts
        lower_bound_dict, upper_bound_dict = {}, {}
        for i,element in enumerate(input_array):
            if element not in lower_bound_dict:
                lower_bound_dict[element] = i
            upper_bound_dict[element] = i

        # Create Elements Info Dict
        elements_info = {}
        for element in inclusive_range(min_element, max_element):
            elements_info[element] = {
                'lower_bound': lower_bound_dict.get(element, -1),
                'upper_bound': upper_bound_dict.get(element, -1),
            }

        # Add to Binary Search Inputs
        binary_search_inputs.append({
            # 'num': num,
            'input_array': input_array,
            'elements_info': elements_info,
        })

    return binary_search_inputs

def get_cumulative_binary_search_inputs(max_array_count):
    cumulative_binary_search_inputs = []
    for array_count in inclusive_range(0, max_array_count):
        for d in get_binary_search_inputs(array_count): cumulative_binary_search_inputs.append(d)
    return cumulative_binary_search_inputs

def run_test_suite(func, max_array_count):
    cumulative_binary_search_inputs = get_cumulative_binary_search_inputs(max_array_count)
    tests = []
    for binary_search_input in cumulative_binary_search_inputs:
        input_array = binary_search_input['input_array']
        for element_to_search, element_info in binary_search_input['elements_info'].items():
            lower_bound = element_info['lower_bound']
            upper_bound = element_info['upper_bound']
            count_in_input_array = upper_bound - lower_bound + 1 if lower_bound != -1 else 0
            try:
                output_idx = func(input_array, element_to_search)
                idx_incorrect = bool_to_num(not(lower_bound <= output_idx <= upper_bound))
                idx_mismatches_lower_bound = bool_to_num(not(output_idx == lower_bound))
                idx_mismatches_upper_bound = bool_to_num(not(output_idx == upper_bound))
                exception_str = None
            except Exception as e:
                output_idx = "NA"
                idx_incorrect = bool_to_num(True)
                idx_mismatches_lower_bound = bool_to_num(True)
                idx_mismatches_upper_bound = bool_to_num(True)
                exception_str = str(e)
            idx_range = str(lower_bound)
            if lower_bound != upper_bound:
                idx_range = f"{lower_bound}, {upper_bound}"
            comment_list = []
            if output_idx != lower_bound:
                comment_list.append(f'Lower Bound Mismatch: output_idx `{output_idx}` != lower_bound `{lower_bound}`')
            if output_idx != upper_bound:
                comment_list.append(f'Upper Bound Mismatch: output_idx `{output_idx}` != upper_bound `{upper_bound}`')
            comments = '\n'.join(comment_list)
            tests.append({
                'input_array': input_array,
                'element_to_search': element_to_search,
                'count_in_input_array': count_in_input_array,
                'idx_range': idx_range,
                'output_idx': output_idx,
                'idx_incorrect': idx_incorrect,
                'exception_str': exception_str,
                'comments': comments,
                # 'idx_mismatches_lower_bound': idx_mismatches_lower_bound,
                # 'idx_mismatches_upper_bound': idx_mismatches_upper_bound,
            })
    df_tests = pd.DataFrame(tests)
    return df_tests

def print_diagnostics(df_tests, verbose=False):
    df_tests_failed = df_tests[df_tests['idx_incorrect']==True]
    df_tests_passed = df_tests[df_tests['idx_incorrect']==False]

    len_df_tests = df_tests.shape[0]
    len_df_tests_failed = df_tests_failed.shape[0]
    len_df_tests_passed = df_tests_passed.shape[0]

    print(f'Failed {len_df_tests_failed}/{len_df_tests} tests')
    if verbose:
        print(df_tests_failed.to_string(index=False))
        print()
    print(f'Passed {len_df_tests_passed}/{len_df_tests} tests')
    if verbose:
        print(df_tests_passed.to_string(index=False))
        print()

def update_sheet(df_tests):
    spreadsheet_name = 'BinarySearch'
    worksheet_name = 'test_results'

    from google.colab import auth
    auth.authenticate_user()

    import gspread
    from google.auth import default
    creds, _ = default()

    gc = gspread.authorize(creds)

    sh = gc.open(spreadsheet_name)
    worksheet = sh.worksheet(worksheet_name)

    worksheet.clear()
    # list of lists to push to google sheets
    sheet_entries = [df_tests.columns.values.tolist()] + df_tests.values.tolist()
    for entry in sheet_entries:
        # problem when first element is a list
        entry[0] = str(entry[0])
    worksheet.update(sheet_entries)

df_tests = run_test_suite(func, 10)
print_diagnostics(df_tests)
update_sheet(df_tests)
```
